```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 300          # Number of electrons
width = 10       # Box width
height = 5       # Box height
dt = 0.05        # Time step
```

```python
q = -1
m = 1
Ex = 10
B = 0.5
ax = q * Ex / m

```

Need to change boundaries here. x-dievtion stays same while y-direction changes to accomodate accumulation of charges


```python
x = np.random.uniform(0, width, N)
y = np.random.uniform(0, height, N)

vx = np.random.normal(0, 1, N)
vy = np.random.normal(0, 1, N)


```

```python
drift = []
P = 0.2

for step in range(5000):
    #magnetic acceleration
    ax_mag = q * vy * B / m
    ay_mag = -q * vx * B / m

   
    #accelerate electrons
    vx += (ax + ax_mag) * dt
    vy += (ay_mag) * dt

    #move electrons about
    x += vx * dt
    y += vy * dt

    #periodic boundaries
    x = x % width
    
    # Top wall
    top = y > height
    y[top] = height
    vy[top] *= -1

    # Bottom wall
    bottom = y < 0
    y[bottom] = 0
    vy[bottom] *= -1


    # Collisions
    collision = np.random.rand(N) < P
    vx[collision] = np.random.normal(0, 1, np.sum(collision))
    vy[collision] = np.random.normal(0, 1, np.sum(collision))


    drift.append(np.mean(vx))

    print(np.sum(collision))
```

```python
plt.figure(figsize=(8,4))
plt.scatter(x, y, s=10)
plt.xlim(0, width)
plt.ylim(0, height)
plt.show()
```
